```
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
x = [2,4,6,8,10,12,14,16,18]
x = np.array(x)
y = [1,3,5,7,9,11,13,15,17]
y = np.array(y)*2
X, Y = np.meshgrid(x, y)

Z = np.sin(X) * np.sin(Y)
plt.contourf(X,Y,Z, cmap="tab20c", alpha=0.7)
plt.colorbar()
```
