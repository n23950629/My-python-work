```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
```

# 讀森林大火資料

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/forest-fires/forestfires.csv')
```

```python
df.head()
```

# 刪除不必要欄位

```python
df=df.drop("area",axis=1)
```

```python
df.head()
```

```python
df=df.drop("rain",axis=1)
```

```python
df.head()
```

# 依照日期和 FFMC 排名

```python
df_sorted=df.sort_values(by=["month","day","FFMC"],ascending=False)
```

```python
df_sorted
```

# 依月分統計

```python
df_area=df.groupby('month').sum()
```
