```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=np.arange(0,720)
y=np.cos(x*np.pi/180.0)
plt.plot(x,y)
plt.xlabel("X axis")            #X軸標記
plt.ylabel("Y axis")            #Y軸標記
plt.title("My cos graph")       #標題
plt.plot(x,y,"orange",lw=5)     #線條寬     #顏色   
```
